```python
import torch
from smolagents import TransformersModel

model_name = "Qwen/Qwen2.5-7B-Instruct"

# Envolvemos el modelo de Transformers en un TransformersModel de smolagents
llm_model = TransformersModel(model_id=model_name, device_map="auto", torch_dtype=torch.float16)
```

```python
import os

# Creamos la carpeta docs si no existe
os.makedirs("docs", exist_ok=True)

# Contenido de ejemplo para el documento (en español)
doc1_text = (
    "La Torre Eiffel es una torre de hierro pudelado de 330 metros de altura ubicada en París. "
    "Originalmente medía 324 metros, pero en marzo de 2022 se instaló una nueva antena en la cima, "
    "alcanzando su altura actual de 330 metros. Es uno de los monumentos más visitados del mundo.\n"
)
# Guardamos el documento de ejemplo en un archivo de texto
with open("docs/torre_eiffel.txt", "w", encoding="utf-8") as f:
    f.write(doc1_text)

print("Documento de ejemplo creado con éxito.")
```

```python
# Segundo documento de ejemplo (opcional)
doc2_text = (
    "El número pi es aproximadamente 3.14159. Es un número irracional famoso en matemáticas, "
    "definido como la relación entre la circunferencia de un círculo y su diámetro.\n"
)
with open("docs/matematicas.txt", "w", encoding="utf-8") as f:
    f.write(doc2_text)

print("Segundo documento de ejemplo creado con éxito.")
```

```python
# Importaciones actualizadas (v0.10.0+)
from llama_index.core import SimpleDirectoryReader, VectorStoreIndex, Settings
from llama_index.legacy import LLMPredictor  # LLMPredictor movido a legacy
from llama_index.embeddings.langchain import LangchainEmbedding
from langchain.embeddings import HuggingFaceEmbeddings
from langchain.llms import HuggingFacePipeline
from transformers import pipeline
import torch

# 1. Cargar documentos desde la carpeta 'docs' (sin cambios)
documents = SimpleDirectoryReader('docs').load_data()
print(f"Se cargaron {len(documents)} documentos.")

# 2. Configurar modelos usando Settings global (reemplaza ServiceContext)
# Pipeline de Hugging Face
generation_pipeline = pipeline(
    "text-generation",
    model="Qwen/Qwen2.5-7B-Instruct",
    device_map="auto",
    # load_in_4bit=True,  # ✅ Activa cuantización 4-bit
    torch_dtype=torch.float16
)

# from accelerate import disk_offload

# # 1. Carga el modelo en CPU
# model = AutoModelForCausalLM.from_pretrained(
#     "Qwen/Qwen2.5-7B-Instruct",
#     device_map="cpu",
#     torch_dtype=torch.float16
# )

# # 2. Offload explícito a disco
# disk_offload(
#     model=model,
#     offload_dir="./model_offload",  # Carpeta temporal
#     force_hooks=True
# )

# Configuración global (nuevo enfoque)
Settings.llm = HuggingFacePipeline(pipeline=generation_pipeline)
Settings.embed_model = LangchainEmbedding(
    HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
)

# 3. Construir el índice vectorial (sin ServiceContext)
index = VectorStoreIndex.from_documents(documents)

# 4. Query engine (sin cambios)
query_engine = index.as_query_engine()
print("Índice de documentos construido y listo para consultas.")
```

```python
# Ejemplo de consulta de prueba al índice (pregunta directa sin pasar por el agente)
prueba = query_engine.query("¿Cuál es la altura actual de la Torre Eiffel?")
print(str(prueba))
```

```python
from smolagents import tool

@tool
def consulta_documentos(pregunta: str) -> str:
    """Esta Tool solo sirve para buscar información relacionada con París o el número Pi. Busca en los documentos proporcionados la respuesta a la pregunta dada.

    Args:
        pregunta: La pregunta que se quiere responder usando los documentos. Debe estar relacionada con París o el número Pi.
    """
    # Utiliza el query_engine de LlamaIndex para obtener la respuesta
    respuesta = query_engine.query(pregunta)
    # Convertimos la respuesta a cadena (puede ser objeto Response de LlamaIndex)
    return str(respuesta)

@tool
def calculadora(expresion: str) -> str:
    """Calcula el resultado de una expresión aritmética simple.

    Args:
        expresion: La expresión matemática a calcular.
    """
    # Nota: esta función evalúa la expresión de forma sencilla.
    # Aseguramos que solo contenga dígitos y operadores básicos antes de evaluar.
    import re
    if re.match(r'^[\d+\-*/(). ]+$', expresion):
        try:
            resultado = eval(expresion)
        except Exception as e:
            resultado = f"Error en cálculo: {e}"
    else:
        resultado = "Expresión no válida"
    return str(resultado)
```

```python
from smolagents import ToolCallingAgent, PromptTemplates



system_msg = PromptTemplates(system_prompt="""
Eres un asistente conversacional que cuenta con dos herramientas:

1. consulta_documentos(pregunta: str): úsala **solo** para responder preguntas relacionadas con Francia.
2. calculadora(expresion: str): úsala **solo** para resolver expresiones matemáticas.

Para cualquier otra pregunta de conocimiento general, RESPONDE DIRECTAMENTE sin llamar a ninguna herramienta.
    """
)

agent = ToolCallingAgent(
    model=llm_model,
    tools=[consulta_documentos, calculadora],
    # prompt_templates=system_msg
)
```

```python
# Ejemplo 1: Pregunta sobre los documentos (altura de la Torre Eiffel)
pregunta1 = "¿Cuál es la altura actual de la Torre Eiffel?"
respuesta1 = agent.run(pregunta1)
print(f"**Usuario:** {pregunta1}\n**Agente:** {respuesta1}\n")

# Ejemplo 2: Pregunta que requiere uso de la calculadora
pregunta2 = "¿Cuál es el resultado de 523 * 67?"
respuesta2 = agent.run(pregunta2)
print(f"**Usuario:** {pregunta2}\n**Agente:** {respuesta2}\n")

# Ejemplo 3: Pregunta de conocimiento general (sin necesidad de herramienta)
pregunta3 = "¿De qué color es el cielo?"
respuesta3 = agent.run(pregunta3)
print(f"**Usuario:** {pregunta3}\n**Agente:** {respuesta3}\n")
```
